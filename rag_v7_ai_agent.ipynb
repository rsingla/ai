```python
from llama_index.agent.openai import OpenAIAssistantAgent
from llama_index.core import (
    SimpleDirectoryReader,
    VectorStoreIndex,
    StorageContext,
)
from llama_index.vector_stores.supabase import SupabaseVectorStore

from llama_index.core.tools import QueryEngineTool, ToolMetadata
```

```python
reader = SimpleDirectoryReader(input_files=["bookpdf/AtomicHabits.pdf"])
docs = reader.load_data()
for doc in docs:
    doc.id_ = "atomic_habits"
```

```python
#Dont try this in Production system, use enviroment variable using secure vault to load the database connection information. 

vector_store = SupabaseVectorStore(
    postgres_connection_string=(
        "postgresql://postgres:password1234@localhost:5433/rag_vector_store"
    ),
    collection_name="base_demo",
)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
index = VectorStoreIndex.from_documents(docs, storage_context=storage_context)
```
